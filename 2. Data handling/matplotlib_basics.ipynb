### ðŸ§® Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

#### 1. Creating a Bar Chart with Matplotlib

```python
subjects = ["Math", "Physics", "Chemistry"]
marks = [78, 85, 90]
plt.bar(subjects, marks)
plt.title("Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

#### 2. Creating a Pie Chart with Matplotlib

```python
apps = ['YouTube', 'WhatsApp', 'Instagram', 'Study App', 'Games']
minutes = [60, 40, 50, 90, 30]

# Pie Chart
plt.pie(minutes, labels=apps, autopct='%1.1f%%', startangle=0)
plt.title('App Usage Distribution')
plt.axis('equal')
plt.show()

```

#### 3. Creating a Line Chart with Matplotlib

```python
# Line Chart: Daily screen time
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
screen_time = [3.5, 4, 2.5, 4.5, 5, 6, 4]

plt.plot(days, screen_time, marker='o', linestyle='-', color='green')
plt.title('Screen Time Trend Over a Week')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```

#### 4. Creating a Histogram for Data Distribution

```python
# Histogram: Session lengths (in minutes)
sessions = [30, 45, 20, 50, 25, 60, 40, 35, 20, 55]

plt.hist(sessions, bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Session Lengths')
plt.xlabel('Session Length (minutes)')
plt.ylabel('Frequency')
plt.show()

```

## ðŸ§ Seaborn for Enhanced Visualizations

```python
import seaborn as sns
import pandas as pd
```

#### 1. Barplot using Seaborn

```python
data = {
    "Subject": ["Math", "Physics", "Chemistry"],
    "Average": [78, 85, 90]
}
df = pd.DataFrame(data)
sns.barplot(x="Subject", y="Average", data=df)
plt.title("Average Subject Scores")
plt.show()
```

#### 2. Visualize Correlations with a Heatmap

```python
marks = {
    "Math": [78, 85, 82],
    "Physics": [89, 90, 88],
    "Chemistry": [85, 80, 91]
}
df = pd.DataFrame(marks)
sns.heatmap(df, annot=True)
plt.title("Heatmap of Student Scores")
plt.show()
```
